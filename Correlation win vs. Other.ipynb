```python
import pandas as pd
import numpy as np
```

```python
lol_df = pd.read_csv("LOL data.csv")
lol_df.head()
```

```python
lol_df = pd.get_dummies(lol_df,
                          columns=['Lane','FirstBloodAssist', 'FirstBloodKill', 'FirstTowerKill', 'FirstTowerAssist','rank'])
```

```python
lol_df['Win'] = np.where(lol_df['Win/Lose'] == True, 1, 0) 
lol_df = lol_df.drop(['Win/Lose', 'ChampionName'], axis = 1)
```

```python
lol_df = lol_df.drop(['Unnamed: 0'], axis = 1) 
```

```python
import seaborn as sn
import matplotlib as plt
```

```python
corrMatrix = lol_df.corr()
sn.heatmap(pd.DataFrame(corrMatrix['Win']), annot=True)
plt.rcParams["figure.figsize"] = (25,20)
```
